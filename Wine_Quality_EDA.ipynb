```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('WineQT.csv')
```

```python
data.info()#Veri seti tamdır. Eksik veri seti yoktur.
```

```python
data.describe()
```

```python
data.drop('Id', axis = 1, inplace = True)
```

```python
group_quality = data.groupby('quality').mean()
group_quality
```

```python
group_quality.plot(kind = 'bar')
plt.xlabel('Quality')
plt.ylabel('Mean')
plt.title('Quality Mean')
plt.show()
```

```python
sns.set_style("darkgrid")
plt.figure(figsize = (6, 5))
sns.countplot(x = "quality", data = data)
plt.xlabel("Quality")
plt.ylabel('Count')
plt.title('Quality Count')
plt.show()
```

```python
corr = data.corr(numeric_only = True)
corr
```

```python
plt.figure(figsize = (10, 7))
sns.heatmap(corr, annot = True, cmap = 'coolwarm', fmt = '.3f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
mask = np.triu(np.ones_like(data.corr(numeric_only = True), dtype = bool))
plt.figure(figsize = (10, 7))
sns.heatmap(corr, annot = True, cmap = 'coolwarm', fmt = '.3f', mask = mask)
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize = (10, 7))
sns.boxplot(x = 'quality', y = 'alcohol', data = data)
plt.xlabel('Quality')
plt.ylabel('Alchohol')
plt.title('Quality vs Alchohol')
plt.show()
```

```python
plt.figure(figsize = (10, 7))
sns.scatterplot(x = 'fixed acidity', y = 'density', data = data)
plt.xlabel('Fixed Acidity')
plt.ylabel('Density')
plt.title('Fixed Acidity vs Density')
plt.show()
```

```python
columns = data.columns
(fig, ax) = plt.subplots(4, 4, figsize = (16, 15))
ax = ax.flatten()

for i, column in enumerate(columns):
    sns.kdeplot(
        data = data,
        x = column,
        hue = data.quality,
        ax = ax[i]
    )
    ax[i].set_title(f'{column} Distribution')
    ax[i].set_xlabel(None)


for i in range(i + 1, len(ax)):
    ax[i].axis("off")
```
